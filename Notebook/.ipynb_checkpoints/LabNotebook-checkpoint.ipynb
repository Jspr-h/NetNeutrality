##### Member 5:
Net Neutrality (NN) is about the freedom of traffic for different internet services (search engines, social media, streaming services and so on). Historically the internet has been unregulated, and the Net Neutrality acts’ main purpose is to keep is the way, by regulating it. How so? The regulations imposed on the internet service providers by the NN regulations means that they cannot favour other companies and make deals about what traffic should have ‘priority’ on the internet.

Basically, the people who invest in, and build the ‘’roads’ (Internet Service Providers), wants to be able to determine the rules of the road, which has historically been illegal. This means that right now the internet is ‘free’ to the consumers who can choose freely between services and expect their ISP’s to not differ between them. This however means that the ISP’s cannot profit from the work they put into creating the infrastructure, and they cannot sell this infrastructure (the roads) to companies providing services that require internet.
It is interesting that suddenly the service providers want to compete on more complex parametres than price (who provides the best traffic for streaming?). The market is currently ‘free’ because it is not a liberated market. The different companies acting on this market, are right now not able to compete on differing services.

3. Why are the companies that stand to be regulated (Netflix, Google, Facebook) not doing more to fight this? They seem to have extensive lobbyism power.
